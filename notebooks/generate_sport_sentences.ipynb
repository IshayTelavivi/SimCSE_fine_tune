```python
def get_random_sport():
    """Get a random sport type"""
    return random.choice(list(athletes.keys()))

def generate_sentence():
    """Generate a random sports headline sentence with sport consistency"""
    sport = get_random_sport()

    patterns = [
        # Pattern 1: [Player] scored [X] [achievement] in [team] [outcome]
        lambda: f"{random.choice(athletes[sport])} scored {random.randint(15, 50)} {random.choice(achievements[sport])} in {random.choice(teams[sport])} {random.choice(outcomes)}",

        # Pattern 2: [Team] defeats [Team] [score]
        lambda: generate_team_vs_team(sport),

        # Pattern 3: [Player] wins [competition] [context]
        lambda: f"{random.choice(athletes[sport])} wins {random.choice(competitions[sport])} {random.choice(['final', 'championship', 'title', 'tournament', 'match', 'series'])}",

        # Pattern 4: [Team] win [competition] in [context]
        lambda: f"{random.choice(teams[sport])} win {random.choice(competitions[sport])} in {random.choice(contexts[sport])}",

        # Pattern 5: [Player] [action] for [team]
        lambda: f"{random.choice(athletes[sport])} {random.choice(actions[sport])} for {random.choice(teams[sport])}",

        # Pattern 6: [Player] leads [team] to [X]-[X] victory over [team]
        lambda: generate_team_victory_with_player(sport),

        # Pattern 7: [Team] clinch [competition] with [context] win
        lambda: f"{random.choice(teams[sport])} clinch {random.choice(['playoff spot', 'division title', 'championship', 'title', 'trophy', 'qualification'])} with {random.choice(['dominant', 'crucial', 'decisive', 'thrilling', 'commanding', 'stunning'])} win",

        # Pattern 8: [Player] breaks record with [X] [achievement] in [context]
        lambda: f"{random.choice(athletes[sport])} breaks record with {random.randint(25, 60)} {random.choice(achievements[sport])} in {random.choice(['season opener', 'playoff game', 'championship final', 'derby match', 'crucial win', 'historic match'])}",

        # Pattern 9: [Player] dominates [competition] with [achievement] performance
        lambda: f"{random.choice(athletes[sport])} dominates {random.choice(competitions[sport])} with {random.choice(['outstanding', 'stellar', 'magnificent', 'incredible', 'phenomenal'])} performance",

        # Pattern 10: [Team] advance to [competition] after [context] victory
        lambda: f"{random.choice(teams[sport])} advance to {random.choice(competitions[sport])} after {random.choice(['hard-fought', 'convincing', 'dramatic', 'comfortable', 'narrow'])} victory"
    ]

    return random.choice(patterns)()

def generate_team_vs_team(sport):
    """Generate team vs team with score for specific sport"""
    team_list = teams[sport]
    team1 = random.choice(team_list)
    team2 = random.choice([t for t in team_list if t != team1])

    if sport == 'basketball':
        score1 = random.randint(85, 130)
        score2 = random.randint(75, 125)
    elif sport == 'soccer':
        score1 = random.randint(1, 5)
        score2 = random.randint(0, 4)
    elif sport == 'football':
        score1 = random.randint(14, 45)
        score2 = random.randint(7, 42)
    else:  # tennis
        sets = random.choice(["3-0", "3-1", "3-2", "2-0", "2-1"])
        return f"{team1} defeats {team2} {sets}"

    return f"{team1} defeats {team2} {score1}-{score2}"

def generate_team_victory_with_player(sport):
    """Generate player leads team to victory for specific sport"""
    player = random.choice(athletes[sport])
    team_list = teams[sport]
    team1 = random.choice(team_list)
    team2 = random.choice([t for t in team_list if t != team1])

    if sport == 'basketball':
        score1 = random.randint(95, 125)
        score2 = random.randint(85, 120)
    elif sport == 'soccer':
        score1 = random.randint(2, 6)
        score2 = random.randint(0, score1 - 1)
    elif sport == 'football':
        score1 = random.randint(21, 42)
        score2 = random.randint(14, 35)
    else:  # tennis
        sets = random.choice(["3-1", "3-2", "2-0", "2-1"])
        return f"{player} leads {team1} to {sets} victory over {team2}"

    return f"{player} leads {team1} to {score1}-{score2} victory over {team2}"

def generate_dataset(num_sentences=5000):
    """Generate unique sports headlines dataset"""
    print(f"Generating {num_sentences} unique sports headlines...")

    sentences = set()
    attempts = 0
    max_attempts = num_sentences * 3  # Prevent infinite loop

    while len(sentences) < num_sentences and attempts < max_attempts:
        sentence = generate_sentence()
        sentences.add(sentence)
        attempts += 1

        if len(sentences) % 500 == 0:
            print(f"Generated {len(sentences)} unique sentences...")

    sentences_list = list(sentences)
    print(f"Successfully generated {len(sentences_list)} unique sentences!")
    return sentences_list

def save_as_json(sentences, filename="sports_headlines_5000.json"):
    """Save dataset as JSON file"""
    dataset = {
        "metadata": {
            "total_sentences": len(sentences),
            "description": "Sports news headlines dataset for LLM fine-tuning",
            "patterns_included": [
                "[Player] scored [X] [achievement] in [team] [outcome]",
                "[Team] defeats [Team] [score]",
                "[Player] wins [competition] [context]",
                "[Team] win [competition] in [context]",
                "[Player] [action] for [team]",
                "[Player] leads [team] to [X]-[X] victory over [team]",
                "[Team] clinch [competition] with [context] win",
                "[Player] breaks record with [X] [achievement] in [context]",
                "[Player] dominates [competition] with [performance] performance",
                "[Team] advance to [competition] after [context] victory"
            ],
            "sports_covered": ["Basketball", "Soccer/Football", "American Football", "Tennis"],
            "athletes_per_sport": {sport: len(athlete_list) for sport, athlete_list in athletes.items()},
            "teams_per_sport": {sport: len(team_list) for sport, team_list in teams.items()},
            "generated_date": datetime.now().isoformat()
        },
        "sentences": [{"id": i+1, "text": sentence} for i, sentence in enumerate(sentences)]
    }

    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(dataset, f, indent=2, ensure_ascii=False)

    print(f"JSON dataset saved as: {filename}")

def save_as_csv(sentences, filename="sports_headlines_5000.csv"):
    """Save dataset as CSV file"""
    with open(filename, 'w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow(["id", "text"])  # Header
        for i, sentence in enumerate(sentences):
            writer.writerow([i+1, sentence])

    print(f"CSV dataset saved as: {filename}")
```
